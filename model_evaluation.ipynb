The following notebook loads the trained model from a google drive folder and perofrmas the model evaluation. 

```python
import numpy as np
import pickle

from google.colab import drive

from tensorflow import keras
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
drive.mount('/content/gdrive')
```

```python
model = keras.models.load_model('gdrive/MyDrive/Computer_Vision/model')
```

```python
with open('gdrive/MyDrive/Computer_Vision/dataset.pickle', 'rb') as f:
    dataset = pickle.load(f)
```

```python
data = dataset['data']
labels = dataset['labels']
lb = LabelBinarizer()
labels = lb.fit_transform(labels)
new_labels = np.zeros((len(labels), 3))
new_labels[:,:2] = labels[:,:2]
new_labels[:,2] = np.sum(labels[:, 2:], axis=1)
labels = new_labels
```

```python
# Split of data into train and test
(trainX, testX, trainY, testY) = train_test_split(data, labels,
    test_size=0.20, stratify=labels, random_state=42)
```

```python
predictedY = model.predict(testX)
```

```python
predictedY_cm = np.argmax(predictedY, axis = 1)
testY_cm=np.argmax(testY, axis=1)
```

```python
confusion_matrix(testY_cm, predictedY_cm, normalize='pred')
```
